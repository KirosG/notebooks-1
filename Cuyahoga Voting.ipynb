I was curious about the *rate* of support, so I put together this notebook to work that out.

```python
from censusreporter_api import get_dataframe, search_places
import pandas as pd
```

```python
df = get_dataframe('b01003','160|05000US39035', include_moe=True)
```

```python
df.sort_values(by='Total',ascending=False).head(10)
```

```python
votes = pd.read_excel('cuyahoga_2012_votes.xls',sheetname=1)
```

```python
sum_cols = list(votes)[2:9]
votes['total'] = votes[sum_cols].sum(axis=1)
```

```python
votes['place'] = votes['Precinct'].apply(lambda x: x.split('-')[0])
```

```python
new_cols = ['unity','precinct','soc','ind','con','lib','dem','rep','gre','write-in','over','under','total','place']
votes = votes.rename(columns=dict(zip(votes.columns,new_cols)))
```

```python
by_place = votes[['place','dem','rep','total']].groupby(by='place').sum()
```

```python
by_place['dem_pct'] = by_place['dem']/by_place['total']
```

```python
by_place.sort_values(by='dem_pct',ascending=False)
```
